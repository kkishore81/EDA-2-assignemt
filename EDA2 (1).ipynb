```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the flight price dataset
df=pd.read_excel('flight_price.xlsx')

# Split the date column into three columns
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]

# Convert the date, month, and year columns to integers
df["Date"] = pd.to_numeric(df["Date"])
df["Month"] = pd.to_numeric(df["Month"])
df["Year"] = pd.to_numeric(df["Year"])

# Calculate average price of flights for each month
df_average_price_by_month = df.groupby("Month")["Price"].mean()
```
